# Data Exploration

## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# matplotlib settings
plt.rcParams['figure.figsize'] = (20, 10)
plt.rcParams['figure.dpi'] = 200
```

## Load Data

```python
df = pd.read_csv("../data/test.csv")
```

## Plot Accelerometer Data

```python
# plot ax, ay, az
plt.plot(df["ax"])
plt.plot(df["ay"])
plt.plot(df["az"])
plt.legend(["ax", "ay", "az"])
plt.show()
```

## Plot Gyroscope Data

```python
# plot gx, gy, gz
plt.plot(df["gx"])
plt.plot(df["gy"])
plt.plot(df["gz"])
plt.legend(["gx", "gy", "gz"])
plt.show()
```

## Plot Magnetometer Data

```python
# plot qx, qy, qz, qw
plt.plot(df["qx"])
plt.plot(df["qy"])
plt.plot(df["qz"])
plt.plot(df["qw"])
plt.legend(["qx", "qy", "qz", "qw"])
plt.show()
```

## Plot Pressure

```python
# plot pressure
plt.plot(df["p"])
plt.legend(["pressure"])
plt.show()
```

## Plot Combined Data

```python
# plot all
df.plot()
plt.show()
```

```python
# plot correlation
corr = df.corr()
# plot correlation matrix
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0, len(df.columns), 1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(df.columns)
ax.set_yticklabels(df.columns)
plt.show()
```
